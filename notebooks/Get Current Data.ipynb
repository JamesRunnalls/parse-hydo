```python
import json
import requests
from datetime import datetime
from bs4 import BeautifulSoup
```

```python
bafu_flow_url = 'https://www.hydrodaten.admin.ch/en/current-situation-table-discharge-and-water-levels.html'
bafu_temp_url = 'https://www.hydrodaten.admin.ch/en/water-temperature-table.html'
zurich_temp_url = 'https://hw.zh.ch/hochwasser/mac/AktWassertemp.HTML'
```

```python
def bafu_unix(string):
    return datetime.strptime(string, "%d.%m.%Y %H:%M").timestamp()
def zurich_unix(string):
    return datetime.strptime(string, "%H:%M%d.%m.%Y").timestamp()
```

Bafu Flow

```python
r = requests.get(bafu_flow_url)
r.encoding = 'utf8'
html_text = r.text
soup = BeautifulSoup(html_text, 'html.parser')
table = soup.find('table', attrs={'class':'table table-bordered'})
table_body = table.find('tbody')
data = []
rows = table_body.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    if len(cols) > 1 and cols[2] != "":
        cols = [cols[0], bafu_unix(cols[2]), cols[3]]
        data.append([ele for ele in cols if ele])
with open('../data/water_flow.json', 'w') as outfile:
    json.dump(data, outfile, ensure_ascii=False)
```

Bafu Temperature

```python
r = requests.get(bafu_temp_url)
r.encoding = 'utf8'
html_text = r.text
soup = BeautifulSoup(html_text, 'html.parser')
table = soup.find('table', attrs={'class':'table table-bordered'})
table_body = table.find('tbody')
data = []
rows = table_body.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    if len(cols) > 1 and cols[2] != "":
        cols = [cols[0], bafu_unix(cols[2]), cols[3]]
        data.append([ele for ele in cols if ele])
with open('../data/water_temperature.json', 'w') as outfile:
    json.dump(data, outfile, ensure_ascii=False)
```

Zurich Temperature

```python
with open('../data/zurich_stations.json') as json_file:
    zs = json.load(json_file)

r = requests.get(zurich_temp_url)
r.encoding = 'utf8'
html_text = r.text
soup = BeautifulSoup(html_text, 'html.parser')
table_body = soup.find('table', attrs={'class':''})
data = []
rows = table_body.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    if len(cols) > 1:
        id = ""
        for s in zs:
            if s["geometry"]["properties"]["name"] == cols[0]:
                id = s["id"]
        cols = [id, zurich_unix(cols[2]), ]
        data.append([ele for ele in cols if ele])
```
